```python
import pandas as pd
import numpy as np
```

```python
e_submit = pd.read_csv('.data/Ego+Crash.csv')
t_submit = pd.read_csv('.data/Timing.csv')
w_submit = pd.read_csv('.data/Weather.csv')
```

```python
submit = pd.read_csv('.data/sample_submission.csv')
```

```python
for i in range(len(submit['label'])):
    if e_submit['label'][i] == 0:
        submit['label'][i] = 0
        
    elif e_submit['label'][i] == 1:
        if w_submit['label'][i] == 0:
            if t_submit['label'][i] == 0:
                submit['label'][i] = 1
            else:
                submit['label'][i] = 2
        elif w_submit['label'][i] == 1:
            if t_submit['label'][i] == 0:
                submit['label'][i] = 3
            else:
                submit['label'][i] = 4
        else:
            if t_submit['label'][i] == 0:
                submit['label'][i] = 5
            else:
                submit['label'][i] = 6
                
    else:
        if w_submit['label'][i] == 0:
            if t_submit['label'][i] == 0:
                submit['label'][i] = 7
            else:
                submit['label'][i] = 8
        elif w_submit['label'][i] == 1:
            if t_submit['label'][i] == 0:
                submit['label'][i] = 9
            else:
                submit['label'][i] = 10
        else:
            if t_submit['label'][i] == 0:
                submit['label'][i] = 11
            else:
                submit['label'][i] = 12
```

```python
submit.to_csv('.data/DACON_4th_Submission.csv', index=False)
```

```python
print("predict\n") # predict label 분포
for i in range(13): 
    num = np.count_nonzero(submit['label']==i)
    print("{}: {}, {:.1f}%".format(i, num,num/len(submit)*100))
```
